## **Data Visualization using Seaborn**

```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

### ***why seaborn?***

### **Seaborn vs. Matplotlib: Key Points**

**1. Ease of Use**
   - **Seaborn**: Higher-level interface, simpler commands for complex visualizations.
   - **Matplotlib**: Lower-level control, requires more lines of code.

**2. Aesthetic Appeal**
   - **Seaborn**: Beautiful default styles and color palettes.
   - **Matplotlib**: Highly customizable but less attractive by default.

**3. Statistical Plots**
   - **Seaborn**: Built-in support for statistical plots (e.g., regression, heatmaps).
   - **Matplotlib**: Requires additional coding for similar plots.

**4. Integrations**
   - **Seaborn**: Better integration with Pandas data frames.
   - **Matplotlib**: Integrates well but needs more manual effort.

**5. Complexity**
   - **Seaborn**: Ideal for quick, high-level analysis.
   - **Matplotlib**: Better for detailed, customized visualizations.


### ***Types of function in seaborn***

- **Figure Level** : Works with the entire figure and can create complex visualizations using multiple subplots
- **Axes Level** : Operates on a single set of axes, typically creating a single plot

### ***Main classification of Graphs*** ###

- **Relational Plot**
- **Distribution Plot**
- **Categorical Plot**
- **Regression Plot**
- **Matrix Plot**
- **Multiplots**

### ***Scatterplot()***

```python
#scatterplot
tips = sns.load_dataset('tips')
tips
```

```python
#scatterplot ---> usign axes function
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
#scatter plot using figure level function
sns.relplot(data=tips,x='total_bill',y='tip',kind = 'scatter')
```

```python
#customization
sns.relplot(data=tips,x='total_bill',y='tip',kind = 'scatter',hue='sex',style='smoker',size = 'size')
```

### ***lineplot()***

```python
gap = px.data.gapminder()
gap
```

```python
temp_df = gap[gap['country']=='Nepal']
temp_df
```

```python
# using axes level function
sns.lineplot(data=temp_df,x='year',y='lifeExp')
```

```python
# using relplot
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line')
```

```python
# lets extract data of nepal,india and china
temp_df = gap[gap['country'].isin(['Nepal','China','India'])]
temp_df
```

```python
sns.relplot(kind='line', data = temp_df, x='year',y='lifeExp',hue = 'country')
```

### ***2. Distribution Plot***

- *used for univariate analysis*
- *used to find out the observayion*
- *Range of the observation*
- *Central tendency*
- *is the data bimodel*
- *Are there outliers*

### **plots under distribution plot**
- ***histplot***
- ***kdeplot***
- ***rugplot***

- ***figure level function*** : displot()
- ***axes level function*** : histplot(), kdeplot(), rugplot()

### ***histplot()***

```python
#plotting univariate histogram
sns.displot(data = tips['total_bill'],kind = 'hist') #bins can be defined using bins=
```

```python
#histogram on categorical column
sns.displot(data = tips['day'],kind = 'hist')
```

```python
sns.displot(data = tips,x = 'tip',kind = 'hist',hue = 'sex')
```

```python
#element --> step
sns.displot(data = tips,x = 'tip', kind = 'hist',hue='sex',element = 'step')
```

```python
#faceting using col and row
sns.displot(data = tips,x = 'tip', kind = 'hist',col='sex')
```

### ***kdeplot()***

```python
#not using dicrete bins like histogram
sns.displot(kind = 'kde',data = tips,x = 'total_bill')
```

```python
sns.displot(kind = 'kde',data = tips,x = 'total_bill',hue='sex')
```

I can understand data distribution more clearly for both male and female in kdeplot than in histogram using hue = 'sex'

```python
sns.displot(kind = 'kde',data = tips,x = 'total_bill',hue='sex',fill = True)
```

### ***Rugplot()***

- *plot marginal distribution by drawing ticks along the x and y axes*
- *this function is intended to complement other functions*

```python
sns.kdeplot(data = tips,x = 'total_bill')
sns.rugplot(data = tips,x = 'total_bill')
```

### ***Bivariate histogram***

```python
sns.histplot(data = tips,x='total_bill',y='tip')
```

*here we are visualizing a distribution (frequency of total_bill over tip),the darker area represent there is high total_bill and tip and lighter area represent frequency count is low*
